# Perform hornet detection on an image or video
This notebook allows you to perform detection with our fine-tuned YOLO model on a custom image or video (mp4).

We run the object detector with the following command from AlexeyAB's "**darknet**" github repository 
(https://github.com/AlexeyAB/darknet):

- For images:
    ```
    ./darknet detector test <path/to/obj.data> <path/to/model.cfg> <path/to/model.weights> <path/to/image.jpg>
    ```
- For videos:
    ```
    ./darknet detector demo <path/to/obj.data> <path/to/model.cfg> <path/to/model.weights> -dont_show <path/to/video.mp4> -i 0 -out_filename <path/to/output.avi>
    ```

````python
# define helper functions
import cv2
import matplotlib.pyplot as plt
  
def imShow(path):
    %matplotlib inline

    image = cv2.imread(path)
    height, width = image.shape[:2]
    resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

    fig = plt.gcf()
    fig.set_size_inches(18, 10)
    plt.axis("off")
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
````

````python
# show image
imShow('predictions.jpg')
````
